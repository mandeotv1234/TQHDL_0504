```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('players_22.csv')
df.head(5)
```

```python
# Chọn các cột: sofifa_id, short_name, age, potential, value_eur, wage_eur, height_cm
df = df[['sofifa_id', 'short_name', 'age', 'potential', 'value_eur', 'wage_eur', 'height_cm']]
df.head(5)
```

```python
df['log_value_eur'] = np.log(df['value_eur'])
df['log_wage_eur'] = np.log(df['wage_eur'])
df = df.drop(columns=['value_eur', 'wage_eur'])
df.head()
```

```python
num_cols = df.select_dtypes(include=[np.number])
num_cols = num_cols.drop(columns=['sofifa_id'])
num_cols_stats = num_cols.agg(['mean', 'std', 'min', 'median', 'max'])
num_cols_stats
```

```python
# Tạo figure và axes với 2 hàng, 3 cột
fig, axes = plt.subplots(2, 3, figsize=(20, 10), sharey=True)

# Danh sách các cột cần vẽ
columns = ['age', 'potential', 'height_cm', 'log_value_eur', 'log_wage_eur']

# Vẽ histogram và đường density cho từng cột
for i, col in enumerate(columns):
    row, col_idx = divmod(i, 3)  # Tính hàng và cột
    sns.histplot(num_cols[col], kde=True, ax=axes[row, col_idx], color='blue')
    axes[row, col_idx].set_title(col)
    axes[row, col_idx].set_xlabel('')
    axes[row, col_idx].set_ylabel('')
    axes[row, col_idx].grid(True)  # Thêm lưới

# Ẩn ô trống cuối cùng
axes[1, 2].axis('off')

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

### Nhận xét dữ liệu từ biểu đồ

1. **Age (Tuổi)**:
   - Phân bố tuổi tập trung chủ yếu trong khoảng từ 20 đến 30 tuổi.
   - Số lượng cầu thủ giảm dần khi tuổi tăng, với rất ít cầu thủ trên 35 tuổi.

2. **Potential (Tiềm năng)**:
   - Phân bố tiềm năng tập trung trong khoảng từ 60 đến 80.
   - Rất ít cầu thủ có tiềm năng dưới 60 hoặc trên 90.

3. **Height_cm (Chiều cao)**:
   - Phân bố chiều cao có dạng gần như chuẩn, tập trung chủ yếu trong khoảng từ 170 cm đến 190 cm.
   - Rất ít cầu thủ có chiều cao dưới 160 cm hoặc trên 200 cm.

4. **Log Value (Giá trị cầu thủ - log)**:
   - Phân bố giá trị log có dạng lệch phải, tập trung trong khoảng từ 12 đến 16.
   - Một số ít cầu thủ có giá trị rất cao, tạo ra đuôi dài ở phía bên phải.

5. **Log Wage (Lương cầu thủ - log)**:
   - Phân bố lương log cũng có dạng lệch phải, tập trung trong khoảng từ 6 đến 10.
   - Một số ít cầu thủ có mức lương rất cao, tạo ra đuôi dài ở phía bên phải.

### Tổng quan:
- Dữ liệu cho thấy sự tập trung của các cầu thủ trẻ, có tiềm năng trung bình cao và chiều cao phổ biến.
- Giá trị và lương của cầu thủ có sự chênh lệch lớn, với một số ít cầu thủ có giá trị và lương rất cao.

```python
# Tính ma trận tương quan
correlation_matrix = num_cols.corr()

# Vẽ heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Tương quan giữa các thuộc tính")
plt.show()
```
