```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('players_22.csv')
df.head(5)
```

```python
# Chọn các cột: sofifa_id, short_name, age, potential, value_eur, wage_eur, height_cm
df = df[['sofifa_id', 'short_name', 'age', 'potential', 'value_eur', 'wage_eur', 'height_cm']]
df.head(5)
```

```python
df['log_value_eur'] = np.log(df['value_eur'])
df['log_wage_eur'] = np.log(df['wage_eur'])
df = df.drop(columns=['value_eur', 'wage_eur'])
df.head()
```

```python
num_cols = df.select_dtypes(include=[np.number])
num_cols = num_cols.drop(columns=['sofifa_id'])
num_cols_stats = num_cols.agg(['mean', 'std', 'min', 'median', 'max'])
num_cols_stats
```

## Câu 1: Các thuộc tính số phân phối như thế nào?

```python
# Tạo figure và axes với 2 hàng, 3 cột
fig, axes = plt.subplots(2, 3, figsize=(20, 10), sharey=True)

# Danh sách các cột cần vẽ
columns = ['age', 'potential', 'height_cm', 'log_value_eur', 'log_wage_eur']

# Vẽ histogram và đường density cho từng cột
for i, col in enumerate(columns):
    row, col_idx = divmod(i, 3)  # Tính hàng và cột
    sns.histplot(num_cols[col], kde=True, ax=axes[row, col_idx], color='blue')
    axes[row, col_idx].set_title(col)
    axes[row, col_idx].set_xlabel('')
    axes[row, col_idx].set_ylabel('')
    axes[row, col_idx].grid(True)  # Thêm lưới

# Ẩn ô trống cuối cùng
axes[1, 2].axis('off')

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

### Nhận xét dữ liệu từ biểu đồ

1. **Age (Tuổi)**:
   - Phân bố tuổi tập trung chủ yếu trong khoảng từ 20 đến 30 tuổi.
   - Số lượng cầu thủ giảm dần khi tuổi tăng, với rất ít cầu thủ trên 35 tuổi.

2. **Potential (Tiềm năng)**:
   - Phân bố tiềm năng tập trung trong khoảng từ 60 đến 80.
   - Rất ít cầu thủ có tiềm năng dưới 60 hoặc trên 90.

3. **Height_cm (Chiều cao)**:
   - Phân bố chiều cao có dạng gần như chuẩn, tập trung chủ yếu trong khoảng từ 170 cm đến 190 cm.
   - Rất ít cầu thủ có chiều cao dưới 160 cm hoặc trên 200 cm.

4. **Log Value (Giá trị cầu thủ - log)**:
   - Phân bố giá trị log có dạng lệch phải, tập trung trong khoảng từ 12 đến 16.
   - Một số ít cầu thủ có giá trị rất cao, tạo ra đuôi dài ở phía bên phải.

5. **Log Wage (Lương cầu thủ - log)**:
   - Phân bố lương log cũng có dạng lệch phải, tập trung trong khoảng từ 6 đến 10.
   - Một số ít cầu thủ có mức lương rất cao, tạo ra đuôi dài ở phía bên phải.

### Tổng quan:
- Dữ liệu cho thấy sự tập trung của các cầu thủ trẻ, có tiềm năng trung bình cao và chiều cao phổ biến.
- Giá trị và lương của cầu thủ có sự chênh lệch lớn, với một số ít cầu thủ có giá trị và lương rất cao.

## Câu 2: Mối tương quan giữa các thuộc tính đã chọn như thế nào?

```python
# Tính ma trận tương quan
correlation_matrix = num_cols.corr()

# Vẽ heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Tương quan giữa các thuộc tính")
plt.show()
```

Dựa trên biểu đồ tương quan:

1. Tuổi (age) và tiềm năng (potential) có mối tương quan âm (-0.26), nghĩa là cầu thủ trẻ có tiềm năng cao hơn so với cầu thủ lớn tuổi.

2. Tiềm năng (potential) và một thuộc tính khác (có thể là overall rating) có mối tương quan mạnh (0.83), cho thấy chỉ số tổng thể hiện tại và tiềm năng có liên quan chặt chẽ.

3. Chiều cao (height_cm) không có mối tương quan đáng kể với các thuộc tính khác.

## Câu 3: Chỉ số chiều cao ảnh hưởng như thế nào đến tiềm năng của cầu thủ?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='height_cm', y='potential', alpha=0.6, color='blue')
plt.title('Mối quan hệ giữa chiều cao (height_cm) và tiềm năng (potential)')
plt.xlabel('Chiều cao (height_cm)')
plt.ylabel('Tiềm năng (potential)')
plt.grid(True)
plt.show()
```

## Câu 4: Tuổi của cầu thủ ảnh hưởng như thế nào đến giá trị của cầu thủ?

```python
# Vẽ biểu đồ phân tán để quan sát mối quan hệ giữa age và value_eur
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='log_value_eur', alpha=0.7, color='blue')
plt.title('Mối quan hệ giữa Tuổi (age) và Giá trị cầu thủ (value_eur)', fontsize=14)
plt.xlabel('Tuổi (age)', fontsize=12)
plt.ylabel('Giá trị cầu thủ (value_eur)', fontsize=12)
plt.grid(True)
plt.show()
```

### Nhận xét từ biểu đồ phân tán:

1. **Mối quan hệ giữa tuổi và giá trị cầu thủ:**
   - Giá trị cầu thủ (log_value_eur) đạt đỉnh ở độ tuổi từ **20 đến 27 tuổi**, cho thấy đây là giai đoạn cầu thủ có giá trị cao nhất.
   - Sau độ tuổi **30**, giá trị cầu thủ giảm dần, điều này có thể do sự suy giảm thể lực hoặc phong độ.

2. **Phân bố giá trị:**
   - Ở độ tuổi trẻ (dưới 20), giá trị cầu thủ khá phân tán, có một số cầu thủ trẻ nổi bật với giá trị rất cao.
   - Ở độ tuổi lớn hơn (trên 30), giá trị cầu thủ tập trung ở mức thấp hơn, ít có cầu thủ có giá trị cao.

3. **Kết luận:**
   - Tuổi tác có ảnh hưởng đáng kể đến giá trị cầu thủ. Giai đoạn từ 20-27 tuổi là thời kỳ đỉnh cao, trong khi giá trị giảm dần khi cầu thủ lớn tuổi hơn.
